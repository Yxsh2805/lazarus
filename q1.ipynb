```python
missing=df_temp.isna().mean()*100
missing_features = missing[ missing>1].index.tolist()
missing_features
```

```python
X = df_temp.drop("Wormhole Stability", axis =1)
y = df_temp["Wormhole Stability"]
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")
df_temp["Plasma Density"]=imputer.fit_transform(df_temp[["Plasma Density"]])
```

```python
from sklearn.compose import make_column_transformer

transformer = make_column_transformer((imputer, missing_features), remainder = "passthrough")

transformed_X = transformer.fit_transform(X)
transformed_X
```

```python
import numpy as np
import optuna
from sklearn.ensemble import VotingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold
def remove_outliers_iqr(X, y, factor=1.5):
    Q1 = np.percentile(X, 25, axis=0)  
    Q3 = np.percentile(X, 75, axis=0)  
    IQR = Q3 - Q1  
    lower_bound = Q1 - factor * IQR
    upper_bound = Q3 + factor * IQR
    
    mask = ~((X < lower_bound) | (X > upper_bound)).any(axis=1)  
    return X[mask], y[mask]
def classify_fit(train_acc, val_acc, threshold=0.1):
    if train_acc - val_acc > threshold:
        return "Overfitting"
    elif train_acc < 0.7 and val_acc < 0.7:
        return "Underfitting"
    else:
        return "Well-Fitted"

def evaluate_models(X, y, n_splits=5, n_trials=20):
    X, y = remove_outliers_iqr(X, y)
    y = np.array(y)
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

    results = {}
    best_ensemble = None  # To store the final ensemble model

    for train_idx, val_idx in skf.split(X, y):
        X_train, X_val = X[train_idx], X[val_idx]
        y_train, y_val = y[train_idx], y[val_idx]

        scaler = StandardScaler()
        X_train = scaler.fit_transform(X_train)
        X_val = scaler.transform(X_val)

        def objective_rf(trial, X_train, y_train, X_val, y_val):
            n_estimators = trial.suggest_int('n_estimators', 50, 200)
            max_depth = trial.suggest_int('max_depth', 3, 20)
            model = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, random_state=42)
            model.fit(X_train, y_train)
            return accuracy_score(y_val, model.predict(X_val))

        def objective_lr(trial, X_train, y_train, X_val, y_val):
            C = trial.suggest_float('C', 0.01, 10.0)
            model = LogisticRegression(C=C, solver='liblinear', max_iter=1000)
            model.fit(X_train, y_train)
            return accuracy_score(y_val, model.predict(X_val))
        
        def objective_knn(trial, X_train, y_train, X_val, y_val):
            n_neighbors = trial.suggest_int('n_neighbors', 3, 15)
            model = KNeighborsClassifier(n_neighbors=n_neighbors)
            model.fit(X_train, y_train)
            return accuracy_score(y_val, model.predict(X_val))

        rf_study = optuna.create_study(direction='maximize')
        rf_study.optimize(lambda trial: objective_rf(trial, X_train, y_train, X_val, y_val), n_trials=n_trials)
        best_rf = RandomForestClassifier(**rf_study.best_params, random_state=42)
        best_rf.fit(X_train, y_train)
        
        lr_study = optuna.create_study(direction='maximize')
        lr_study.optimize(lambda trial: objective_lr(trial, X_train, y_train, X_val, y_val), n_trials=n_trials)
        best_lr = LogisticRegression(**lr_study.best_params, solver='liblinear', max_iter=1000)
        best_lr.fit(X_train, y_train)

        knn_study = optuna.create_study(direction='maximize')
        knn_study.optimize(lambda trial: objective_knn(trial, X_train, y_train, X_val, y_val), n_trials=n_trials)
        best_knn = KNeighborsClassifier(**knn_study.best_params)
        best_knn.fit(X_train, y_train)

        ensemble = VotingClassifier(
            estimators=[('random_forest', best_rf), ('logistic_regression', best_lr), ('knn', best_knn)],
            voting='soft'
        )
        ensemble.fit(X_train, y_train)
        best_ensemble = ensemble  # Save the ensemble model

        for model_name, model in zip(["RandomForest", "LogisticRegression", "KNN", "EnsembleModel"],
                                     [best_rf, best_lr, best_knn, ensemble]):
            if model_name not in results:
                results[model_name] = {
                    "Mean Training Accuracy": [], "Mean Validation Accuracy": [], "F1 Score": [], "Confusion Matrix": []
                }
            
            train_acc = accuracy_score(y_train, model.predict(X_train))
            val_acc = accuracy_score(y_val, model.predict(X_val))
            f1 = f1_score(y_val, model.predict(X_val), average="weighted")
            conf_matrix = confusion_matrix(y_val, model.predict(X_val))
            fit_type = classify_fit(train_acc, val_acc)
            
            results[model_name]["Mean Training Accuracy"].append(train_acc)
            results[model_name]["Mean Validation Accuracy"].append(val_acc)
            results[model_name]["F1 Score"].append(f1)
            results[model_name]["Confusion Matrix"].append(conf_matrix)
            results[model_name]["Fit Type"] = fit_type
    
    for model_name in results:
        results[model_name]["Mean Training Accuracy"] = np.mean(results[model_name]["Mean Training Accuracy"])
        results[model_name]["Mean Validation Accuracy"] = np.mean(results[model_name]["Mean Validation Accuracy"])
        results[model_name]["F1 Score"] = np.mean(results[model_name]["F1 Score"])
        results[model_name]["Confusion Matrix"] = np.sum(results[model_name]["Confusion Matrix"], axis=0).tolist()
    
    return results, best_ensemble

# Train models and get the best ensemble
results, ensemble = evaluate_models(transformed_X, y)
print(results)
```
